```python
import glob
from PIL import Image

import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

# Подготовка датасета

Для того, чтобы научить модель распознавать сгенерированные фотографии, нам нужно взять фотографии двух типов
- искусственные фотографии (которые мы как раз и хотим находить)
- обычные фотографии

Для искуственных фотографий возьмем датасет статьи **A Style-Based Generator Architecture for Generative Adversarial Networks** - https://drive.google.com/drive/folders/1LYURI-LwRSj1f-1O4POJOv06b3kfFKi9  
Для настоящих фотографий возьмем датасет **CelebA** - https://drive.google.com/drive/folders/0B7EVK8r0v71pTUZsaXdaSnZBZzg?resourcekey=0-rJlzl934LzC-Xp28GeIBzQ


Каждую фотографию приведем к единому формату, а именно:
- изменим размер до 28х28 пикселей
- переведем в черно-белый формат

```python
IMAGE_SIZE = (28, 28)

def preprocess_image(img):
    image = Image.fromarray(img)
    image = image.resize(IMAGE_SIZE)
    image = image.convert('L')
    
    return np.asarray(image)
```

```python
fake_images = []
for fake_img_path in glob.glob("fake/*.png")[:1000]:
    fake_image = Image.open(fake_img_path)
    fake_images.append(np.asarray(fake_image))
fake_images = np.asarray(fake_images)
```

```python
plt.figure(figsize=(10, 10))
plt.suptitle("Искуственные фотографии", fontsize=14)
for i, image in enumerate(fake_images[:9]):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(image.astype("uint8"), cmap='gray', vmin=0, vmax=255)
    plt.axis("off")
```

```python
real_images = []
for real_img_path in glob.glob("real/*.jpg")[:1000]:
    real_img = Image.open(real_img_path)
    real_images.append(np.asarray(real_img))
real_images = np.asarray(real_images)
```

```python
plt.figure(figsize=(10, 10))
plt.suptitle("Реальные фотографии", fontsize=14)
for i, image in enumerate(real_images[:9]):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(image.astype("uint8"), cmap='gray', vmin=0, vmax=255)
    plt.axis("off")
```

```python
fake_labels = np.zeros(len(fake_images))
real_labels = np.ones(len(real_images))

images = np.concatenate(([preprocess_image(img) for img in fake_images], [preprocess_image(img) for img in real_images]))
labels = np.concatenate((fake_labels, real_labels))

shuffle_idx = np.random.permutation(len(labels))
images = images[shuffle_idx]
labels = labels[shuffle_idx]
```

# Train/Test split

Разобьем наши фотографии на две группы - train и test. На train обучим нашу модель, а на test проверим качество (чтобы понять, что модель не переобучилась)

```python
train_size = int(len(labels) * 0.8)
train_images = images[:train_size]
train_labels = labels[:train_size]

test_images = images[train_size:]
test_labels = labels[train_size:]
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=IMAGE_SIZE),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

# Обучим нашу модель

```python
model.fit(train_images, train_labels, epochs=20)
```

# Проверим качество предсказания модели на test сплите

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)

print('\nTest accuracy:', test_acc)
```
